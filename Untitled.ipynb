```python
import undetected_chromedriver as uc 
import csv
import re
import time
```

```python
handler_list = []

with open("res.csv",encoding='utf-8',errors='ignore') as rs:
    csv_reader = csv.reader(rs)
    for line in csv_reader:
        handler_list.append(line[2].split('@')[0])

print(len(handler_list))
```

```python
modified = []
for handler in handler_list:
    while(len(handler) and handler[0] == '.'):
        handler = handler[1:]
    
    if(handler == ""):
        print("found")
        continue
    modified.append(handler)

print(len(modified))
    
```

```python
driver = uc.Chrome()
```

```python
driver.get("https://login.yahoo.com")
```

```python
create_acc_btn = driver.find_element("xpath",'//a[@id="createacc"]')
create_acc_btn.click()
```

```python
email_field = driver.find_element("xpath",'//input[@name="userId"]')
email_field.click()
email_field.clear()
```

```python
#email_field.send_keys(modified[0])
exists = False
try:
    err_msg = driver.find_element("xpath",'//div[@data-error="messages.IDENTIFIER_EXISTS"]')
    exists = True
    print("Not a Dead mail")
except:
    print("found a dead")

name_field = driver.find_element("xpath",'//input[@id="usernamereg-firstName"]')
name_field.click()
    
```

```python
dead_mails = []
```

```python
err_msg = driver.find_element("xpath",'//div[@id="reg-error-userId"]')
print(len(err_msg.text))
```

```python
for handler in modified:
    email_field = driver.find_element("xpath",'//input[@name="userId"]')
    email_field.click()
    email_field.clear()
    email_field.send_keys(handler)
    time.sleep(2)
    name_field = driver.find_element("xpath",'//input[@id="usernamereg-firstName"]')
    name_field.click()
    time.sleep(3)

    err_msg = driver.find_element("xpath",'//div[@id="reg-error-userId"]')
    print(len(err_msg.text))
    
    if(len(err_msg.text) != 0):
        print("Not a Dead mail")
    else:
        dead_mails.append(f"{handler}@yahoo.com")
        print(f"found a dead {handler}@yahoo.com")

    print(f"summary: len = {len(dead_mails)}")
    time.sleep(2)
    
        
```

```python
filtered = []
for dead in dead_mails:
    if(len(dead) <= 32 and not dead[0].isdigit()):
        filtered.append(dead)
print(len(filtered))
```

```python
print(len(filtered))
actual_dead = []
```

```python
for handler in filtered:
    email_field = driver.find_element("xpath",'//input[@name="userId"]')
    email_field.click()
    email_field.clear()
    email_field.send_keys(handler.split('@')[0])
    time.sleep(2)
    name_field = driver.find_element("xpath",'//input[@id="usernamereg-firstName"]')
    name_field.click()
    time.sleep(3)
    err_msg = driver.find_element("xpath",'//div[@id="reg-error-userId"]')
    
    if(len(err_msg.text) != 0):
        print("Not a Dead mail")
    else:
        actual_dead.append(f"{handler}")
        print(f"found a dead {handler}")

    print(f"summary: len = {len(actual_dead)}")
    time.sleep(2)
    
        
```

```python
with open("deads.txt","w",newline="") as fo:
    for email in actual_dead:
        fo.write(f"{email}\n")
```
